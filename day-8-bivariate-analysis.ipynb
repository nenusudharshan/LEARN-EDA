## ðŸŒŸ What is Bivariate Analysis?

**Bivariate Analysis** is the process of exploring the relationship between **two variables** in a dataset.  
It helps to:
- Understand how one variable influences or is related to another
- Identify correlations and patterns

---

## ðŸ§© Techniques of Bivariate Analysis

### ðŸ”¸ For **Numerical vs Numerical**
1. **Scatter Plot**
2. **Correlation Coefficient**
3. **Heatmap**

### ðŸ”¸ For **Categorical vs Numerical**
1. **Box Plot**
2. **Violin Plot**
3. **Bar Plot (Grouped)**

### ðŸ”¸ For **Categorical vs Categorical**
1. **Crosstab**
2. **Stacked Bar Plot**

---

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.info()
```

```python
#numerical vs category 
sns.scatterplot(x=df['age'],y=df['fare'])
plt.show()
```

```python
#category vs numerical
sns.boxplot(x=df['pclass'],y=df['fare'])
plt.show()
```

```python
#category vs category 
sns.countplot(x=df['sex'],hue=df['survived'])
plt.show()
```

```python
#numerical vs numerical
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
corr = df[numerical_cols].corr()
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
#category vs numerical
sns.violinplot(x=df['pclass'],y=df['age'])
plt.show()
```

```python
sns.kdeplot(df[df['survived'] == 1]['fare'], label="Survived", shade=True)
sns.kdeplot(df[df['survived'] == 0]['fare'], label="Not Survived", shade=True)
plt.legend()
plt.show()
```
