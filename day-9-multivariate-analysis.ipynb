## ðŸŒŸ What is Multivariate Analysis?

**Multivariate Analysis** is the process of analyzing **three or more variables** together to understand relationships and patterns among them.

It helps to:
- Discover **complex interactions** between multiple variables
- Visualize **multi-dimensional data**
- Detect hidden patterns and clusters

---

## ðŸŽ¯ Popular Techniques of Multivariate Analysis

| Method              | Description                                                     |
|--------------------|---------------------------------------------------------------|
| **Pair Plot**      | Visualizes pairwise relationships between variables             |
| **PairGrid**      | Customizable version of Pair Plot                                |
| **JointGrid**     | Combines scatter plot with univariate plots                      |
| **FacetGrid**     | Creates a grid of plots based on categorical variables           |
| **Heatmap**       | Shows correlation between numerical variables in a matrix form   |

---

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')
df.info()
```

```python
# Create PairGrid
g = sns.PairGrid(df, hue="species")
g.map_diag(sns.histplot,kde=True)  # Histogram on the diagonal
g.map_offdiag(sns.scatterplot)  # Scatterplots elsewhere
g.add_legend()

plt.show()
```

```python
#jointgrid
g = sns.JointGrid(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.histplot,kde=True)
plt.show()
```

```python
#facetgrid
g = sns.FacetGrid(df,col='species')
g.map(sns.histplot,'bill_length_mm')
plt.show()
```
