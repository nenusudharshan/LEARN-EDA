## ðŸŒŸ What is Univariate Analysis?

**Univariate Analysis** is the simplest form of data analysis where we analyze **only one variable** at a time.  
It helps to:
- Understand the distribution of data
- Detect patterns, skewness, and outliers
- Summarize the variable using **statistical measures** and **visualizations**

---

## ðŸ§© Techniques of Univariate Analysis

### ðŸ”¸ For **Numerical Data**
1. **Summary Statistics** (Mean, Median, Mode, Standard Deviation)
2. **Visualization:**
   - **Histogram**
   - **Boxplot**
   - **KDE Plot** (Kernel Density Estimation)

### ðŸ”¸ For **Categorical Data**
1. **Frequency Count**
2. **Visualization:**
   - **Bar Plot**
   - **Pie Chart**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
df.info()
```

```python
print(df['age'].isnull().sum())
```

```python

#histogram
sns.histplot(df['age'],bins=20,kde=True)
plt.show()
```

```python
sns.boxplot(x=df['fare'])
plt.show()
```

```python
sns.countplot(x=df['pclass'])
plt.show()
```

```python
sns.countplot(x=df['pclass'])
plt.show()
```

```python
sns.countplot(x=df['sex'])
plt.show()
```

```python
sns.countplot(x=df['embarked'])
plt.show()
```
